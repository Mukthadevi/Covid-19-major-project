```
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
```

```
# Reading the dataset

data = pd.read_csv('/content/drive/My Drive/ DS/complete.csv') 
data
```

```
data.tail()
```

```
# Extracting the recent records

data_latest = data[data['Date']=="2020-08-06"]
data_latest
```

```
data_latest.shape
```

```
# Total positive cases confirmed as of 6/8/2020

total_positive = data_latest['Total Confirmed cases'].sum()

print("Total positive cases confirmed as of 6/8/2020 = ",total_positive)
```

```
# Total number of discharges reported as of 6/8/2020

total_discharged = data_latest['Cured/Discharged/Migrated'].sum()
print("Total discharges reported as of 6/8/2020 = ",total_discharged)
```

```
#Total number of deaths
```

```
 # Plotting

 data_latest = data_latest.sort_values(by=['Total Confirmed cases'],ascending = False)
 plt.figure(figsize=(12,8),dpi= 80)
 plt.bar(data_latest['Name of State / UT'][:7],data_latest['Total Confirmed cases'][:7], align='center',color='Blue')
 plt.ylabel("Number of confirmed cases",size=20,color='Red')
 plt.title("States with maximum confirmed cases",size=20,color='Red')
 plt.show()
```

```
 # plot to show States with maximum deaths

 data_latest = data_latest.sort_values(by=['Death'],ascending=True)
 plt.figure(figsize=(12,8),dpi= 80)
 plt.bar(data_latest['Name of State / UT'][:7],data_latest['Death'][:7], align='center',color='Green')
 plt.ylabel("Number of Deaths",size=20,color='Red')
 plt.title("States with maximum Deaths",size=20,color='Red')
 plt.show()
```

```
population_data = pd.read_excel("/content/drive/MyDrive/ DS/Book1.xlsx")
population_data.head()
```

```
pop_data = data_latest.drop(['New cases','New deaths','New recovered','Latitude','Longitude',],axis=1)
pop_data.head()
```

```
covid_latest = pd.merge(pop_data,population_data,on='Name of State / UT')
covid_latest.head()
```

```
covid_latest['cases_per_million']=(covid_latest['Total Confirmed cases']/covid_latest['Population'])*1000000
covid_latest.sort_values(by='cases_per_million',ascending=False)
covid_latest.head()
```

```
# graph

df = covid_latest[(covid_latest['Total Confirmed cases']>=50000) | (covid_latest['cases_per_million']>=500)]
plt.figure(figsize=(12,18),dpi=80)
plt.scatter(covid_latest['Total Confirmed cases'],covid_latest['cases_per_million'],alpha=0.5)
plt.xlabel('Number of confirmed cases',size=12)
plt.ylabel('Number of cases per million',size=12)
plt.scatter(df['Total Confirmed cases'],df['cases_per_million'],color='Blue')

for i in range(df.shape[0]):
    plt.annotate(df['Name of State / UT'].tolist()[i], xy=(df['Total Confirmed cases'].tolist()[i], df['cases_per_million'].tolist()[i]),
                xytext = (df['Total Confirmed cases'].tolist()[i]+1.0, df['cases_per_million'].tolist()[i]+12.0), size=11)
plt.tight_layout()    
plt.title('Visualization of variations of Covid_19 figures in different Indian states',size=16)
plt.show()
```

```
#Heatmap

plt.figure(figsize = (12,8))
sns.heatmap(covid_latest.corr(), annot=True)
```

```
# Analysis of individual data
# Here I have taken the data of individuals from 27th October to 12th November.

individual_data = pd.read_csv('/content/drive/MyDrive/ DS/raw_data18.csv')
individual_data.head()
```

```
individual_data.isna().sum()
# Here we find huge number of missing data.
```

```
# Grouping the data based on the disticts the cases were confirmed in.

individual_data_district = individual_data.groupby('Detected District')
individual_data_district = individual_data_district['Entry_ID']
individual_data_district.columns = ['count']
individual_data_district.count().sort_values(ascending=False)
```

```
# This shows that from the 27th of oct to 12th Nov, Thiruvananthapuram district recorded highest number of cases followed by Thrissur.
```

```
individual_data_state = individual_data.groupby('Detected State')
individual_data_state = individual_data_state['Entry_ID']
individual_data_state.columns = ['count']
individual_data_state.count().sort_values(ascending=False).head()
```

```
# Similarlly, Uttar Pradesh recorded highest number of cases between 27th Oct to 12th Nov
```

```
individual_data_gender = individual_data.groupby('Gender')
individual_data_gender = individual_data_gender['Entry_ID']
individual_data_gender.columns = ['count']
#individual_data_gender = pd.DataFrame(individual_data_gender.size().reset_index(name = "count"))
individual_data_gender.count() 
```

```
individual_data_gender = individual_data.groupby('Gender')
individual_data_gender = pd.DataFrame(individual_data_gender.size().reset_index(name = "count"))
individual_data_gender.head()
plt.figure(figsize=(10,6), dpi=80)
barlist = plt.bar(individual_data_gender['Gender'], individual_data_gender['count'], align = 'center', color='Blue', alpha=0.3)
barlist[1].set_color('r')
plt.ylabel('Count', size=12)
plt.title('Count on the basis of gender', size=20)
plt.show()
```

```
# COVID_19 WORLD ANNALYSIS

covid_world = pd.read_csv("/content/drive/MyDrive/ DS/owid-covid-data.csv")
covid_world.head()
```

```
covid_world1 = covid_world[covid_world['date']=='2020-09-01']
covid_world1.head()
```

```
# Total Cases Confirmed as of 09/01/2020

tot_pos = covid_world1['total_cases'].sum()
print('Total Cases confirmed are',tot_pos)
```

```
# Total Deaths as of 09/01/2020

tot_death = covid_world1['total_deaths'].sum()
print("Total deaths reported are",tot_death)
```

```
covid_world1 =  covid_world1.sort_values(by='total_cases',ascending=False)
plt.figure(figsize=(12,8),dpi=80)
plt.bar(covid_world1['location'][1:5],covid_world1['total_cases'][1:5],align='center',color='Blue')
plt.ylabel('Number of Confirmed cases',size=14)
plt.title('Countries with maximum confirmed cases',size=16)
plt.show()
```

```
# Analysis based on deaths

covid_world1 = covid_world1.sort_values(by='total_deaths',ascending=False)
plt.figure(figsize=(12,8),dpi=80)
plt.bar(covid_world1['location'][1:5],covid_world1['total_deaths'][1:5],align='center',color='pink')
plt.ylabel("Number of deaths reported",size=12)
plt.title('Countries with maximum death reports',size=16)
plt.show()
```
